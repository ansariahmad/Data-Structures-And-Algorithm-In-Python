# Palindrome Partitioning

```python
def partition(s):
    print(s)
    if len(s) == 1:
        print("end", s)
        return
    
    for idx in range(len(s)):
        print(idx)
        partition(s[0:idx+1])
        partition(s[idx+1:])
```

```python
s = "aabb"
```

```python
for idx in range(0, len(s)):
    print(s[0:idx+1], s[idx+1:])
```

```python
partition(s)
```
